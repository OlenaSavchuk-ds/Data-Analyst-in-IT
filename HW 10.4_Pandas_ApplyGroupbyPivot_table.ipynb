–¶—è —Ç–∞–±–ª–∏—Ü—è –¥–æ–ø–æ–º–æ–∂–µ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –≤ —è–∫—ñ–π —Ñ—ñ–ª—ñ—ó –∫—É–ø—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∫–æ–∂–Ω–æ—ó –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.

 –ü–∏—Ç–∞–Ω–Ω—è –º–æ–∂—É—Ç—å –±—É—Ç–∏ –±—É–¥—å-—è–∫–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ, –∞–ª–µ —ó—Ö –º–∞—î –±—É—Ç–∏ 3. –ö–æ–∂–Ω–µ –ø–∏—Ç–∞–Ω–Ω—è –æ—Ü—ñ–Ω—é—î—Ç—å—Å—è —è–∫ 1 –∑–∞–≤–¥–∞–Ω–Ω—è. –ë–µ–∑ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –î–ó –Ω–µ –ø—Ä–∏–π–º–∞—î—Ç—å—Å—è. –Ø–∫—â–æ —î –ø–∏—Ç–∞–Ω–Ω—è —â–æ–¥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è - –ø–∏—à—ñ—Ç—å —É —á–∞—Ç üôå


**–ü–æ-–ø–µ—Ä—à–µ**,

*—Å–ø—Ä–æ–±—É—î–º–æ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –∞ —è–∫—ñ –∫–ª—ñ—î–Ω—Ç–∏ (—á–æ–ª–æ–≤—ñ–∫–∏, –∂—ñ–Ω–∫–∏) –ø—Ä–∏–Ω–æ—Å—è—Ç—å –Ω–∞–π–±—ñ–ª—å—à–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –≤ —Ä–æ–∑—Ä—ñ–∑—ñ –º—ñ—Å—Ç —Ñ—ñ–ª—ñ–π*

**—Ç–∞ –≤—ñ–¥–ø–æ–≤—ñ—Å—Ç–∏ –Ω–∞ –ø–µ—Ä—à–µ –ø–∏—Ç–∞–Ω–Ω—è**,
*—è–∫–∏–π –≤–Ω–µ—Å–æ–∫ —É –ø—Ä–∏–±—É—Ç–æ–∫ —Ä–æ–±–ª—è—Ç—å —Ä—ñ–∑–Ω—ñ –≥—Ä—É–ø–∏ –ø–æ–∫—É–ø—Ü—ñ–≤ —É –∫–æ–∂–Ω–æ–º—É –º—ñ—Å—Ç—ñ?*

```python
gender_sales_cities = pd.pivot_table(df,
                                    index='City',
                                    columns='Gender',
                                    values='gross income',
                                    aggfunc='sum'
                                    ).round(2)
gender_totals = gender_sales_cities.sum()                             # –ó–∞–≥–∞–ª—å–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –¥–ª—è Female —Ç–∞ Male
gender_part = (gender_totals / gender_totals.sum() * 100).round(0)    # –ß–∞—Å—Ç–∫–∞ (%) –≤—ñ–¥ –∑–∞–≥–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∏–±—É—Ç–∫—É

gender_sales_cities.loc['Part'] = gender_part
gender_sales_cities
```

–í Naypyitaw –∫–ª—ñ—î–Ω—Ç–∏ –æ—Ö–æ—á—ñ—à–µ –ø–ª–∞—Ç—è—Ç—å –≤–µ–ª–∏–∫–∏–º–∏ —Å—É–º–∞–º–∏ –µ–ª–µ–∫—Ç—Ä–æ–Ω–Ω–∏–º–∏ –≥–∞–º–∞–Ω—Ü—è–º–∏.

**–ë—ñ–∑–Ω–µ—Å-—Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—ó:**

1. –°—Ç–∏–º—É–ª—é–≤–∞—Ç–∏ –æ–ø–ª–∞—Ç—É –∫–∞—Ä—Ç–∫–æ—é —É Yangon —Ç–∞ Mandalay (—Ä–æ–∑—ñ–≥—Ä–∞—à —Ç–æ–≤–∞—Ä—ñ–≤ —Ç—ñ–ª—å–∫–∏ —Å–µ—Ä–µ–¥ –ø–æ–∫—É–ø—Ü—ñ–≤ —ñ–∑ –∫–∞—Ä—Ç–∫–∞–º–∏, –ø—Ä—ñ–æ—Ä–∏—Ç–µ—Ç–Ω–∞ –∫–∞—Å–∞)
2. –†–æ–∑–≤–∏–≤–∞—Ç–∏ —Ç–∞ –ø—Ä–æ—Å—É–≤–∞—Ç–∏ E-wallet —É Naypyitaw (–Ω–∞–π–≤–∏—â–∏–π —Å–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫) (–¥–æ–¥–∞—Ç–∏ –±–æ–Ω—É—Å–Ω—ñ –±–∞–ª–∏ –∑–∞ –æ–ø–ª–∞—Ç—É e-wallet, –∑—Ä–æ–±–∏—Ç–∏ QR-–æ–ø–ª–∞—Ç—É —à–≤–∏–¥—à–æ—é, –∑—Ä–æ–±–∏—Ç–∏ –ø—Ä–æ–º–æ)
3. –°—Ç–∏–º—É–ª—é–≤–∞–Ω–Ω—è –ø–µ—Ä–µ—Ö—ñ–¥ –∑ Cash –Ω–∞ —ñ–Ω—à—ñ —Ç–∏–ø–∏ –æ–ø–ª–∞—Ç

**–¢–æ–º—É —â–æ –¥–∞–Ω—ñ –ø–æ–∫–∞–∑—É—é—Ç—å, —â–æ —Ç–∏–ø –æ–ø–ª–∞—Ç–∏ –ø—Ä—è–º–æ –≤–ø–ª–∏–≤–∞—î –Ω–∞ —Ä–æ–∑–º—ñ—Ä —á–µ–∫—É, –Ω–∞ –ø—Ä–∏–±—É—Ç–æ–∫ –º–∞–≥–∞–∑–∏–Ω—ñ–≤!!!**

**–¢—Ä–µ—Ç—î –ø–∏—Ç–∞–Ω–Ω—è:**

—É —è–∫–∏—Ö –≥–æ–¥–∏–Ω–∞—Ö –¥–Ω—è –ø—Ä–∏–±—É—Ç–æ–∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∏–π —ñ —á–∏ —Å–ø—ñ–≤–ø–∞–¥–∞—î —Ü–µ –∑ –ø—ñ–∫–æ–≤–∏–º–∏ –≥–æ–¥–∏–Ω–∞–º–∏ –ø–æ–∫—É–ø–æ–∫?
–¶–µ –¥–æ–∑–≤–æ–ª–∏—Ç—å:
- –æ–ø—Ç–∏–º—ñ–∑—É–≤–∞—Ç–∏ –≥—Ä–∞—Ñ—ñ–∫ –ø–µ—Ä—Å–æ–Ω–∞–ª—É (–∫–∞—Å–∏ —á–∏ —Å–∫–ª–∞–¥–∏);
- –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –≥–æ–¥–∏–Ω–∏ –¥–ª—è –∞–∫—Ü—ñ–π, –ø—Ä–æ–º–æ —á–∏ –¥–µ–≥—É—Å—Ç–∞—Ü—ñ–π;
- –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, —É –ø—ñ–∫–æ–≤—ñ –≥–æ–¥–∏–Ω–∏ –ª—é–¥–∏ –∫—É–ø—É—é—Ç—å –º–µ–Ω—à–µ –¥–æ—Ä–æ–≥–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤ —ñ —á–∏ –ø–æ—Ç—Ä—ñ–±–Ω–∞ –∑–º—ñ–Ω–∞ –≤–∏–∫–ª–∞–¥–∫–∏ –ø–æ–ª–∏—Ü—å?

```python
def get_hour(t):
    return int(t.split(':')[0])

df['Hour'] = df['Time'].apply(get_hour)

# —Å—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –ø–æ –≥–æ–¥–∏–Ω–∞–º
income_per_hour = df.groupby('Hour')['gross income'].sum().round(2)
income_per_hour
```

```python
# –∫—ñ–ª—å–∫—ñ—Å—Ç—å —ñ–Ω–≤–æ–π—Å—ñ–≤ –ø–æ –≥–æ–¥–∏–Ω–∞–º (–ø—ñ–∫–æ–≤—ñ –≥–æ–¥–∏–Ω–∏)
transactions_per_hour = df.groupby('Hour')['Invoice ID'].count()
transactions_per_hour
```

```python
# –û–±'—î–¥–Ω–∞–Ω–∞ —Ç–∞–±–ª–∏—Ü—è
hour_analysis = pd.DataFrame({
                            'trans_count': transactions_per_hour,
                            'gross_income_sum': income_per_hour
                            })
hour_analysis = hour_analysis.sort_values(by='gross_income_sum', ascending=False)
hour_analysis
```
