1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–∫–Ω—Ü—ñ—é `print_message`, —è–∫–∞ –≤–∏–≤–æ–¥–∏—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω "–Ø –ø—Ä–æ–≥—Ä–∞–º—É—é –Ω–∞ Pythonüí™" —ñ –≤–∏–∫–ª–∏—á—ñ—Ç—å —Ü—é —Ñ—É–∫–Ω—Ü—ñ—é 5 —Ä–∞–∑—ñ–≤ (–Ω–∞–π–∫—Ä–∞—â–µ —Ü–µ –∑—Ä–æ–±–∏—Ç–∏ –≤ —Ü–∏–∫–ª—ñ).

  **–£–≤–∞–≥–∞**: —Ñ—É–∫–Ω—Ü—ñ—è –º–∞—î —Ä–æ–±–∏—Ç–∏ —á—ñ—Ç–∫–æ –ª–∏—à–µ –∑–∞–∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª. –ù–∞—è–≤–Ω—ñ—Å—Ç—å –¥–æ–¥–∞—Ç–∫–æ–≤–æ–≥–æ —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É —ñ –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤, —è–∫—ñ –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω—ñ –¥–ª—è –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –±–µ–∑–ø–æ—Å–µ—Ä–µ–¥–Ω—å–æ –∑–∞–≤–¥–∞–Ω–Ω—è, –±—É–¥–µ –≤–≤–∞–∂–∞—Ç–∏—Å—å –ø–æ–º–∏–ª–∫–æ—é.

````python
def print_message():
    print("–Ø –ø—Ä–æ–≥—Ä–∞–º—É—é –Ω–∞ Pythonüí™")
````

````python
i = 1

while i <= 5:
  print_message()
  i += 1
````

2. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `calculation` —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ 4 –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É int - `a`, `b`, `c`, `d` —ñ –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç–∞–∫–æ—ó —Ñ–æ—Ä–º—É–ª–∏: $ (a + b - c) * d $.

  –í–∏–∫–ª–∏—á—Ç–µ —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –Ω–∞–±–æ—Ä—ñ –∞—Ä–≥—É–º–µ–Ω—Ç—ñ–≤ —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω:
  ```
  a = 16
  b = 4
  c = 9
  d = 3
  ```

````python
def calculation(a,b,c,d):
    return (a + b - c) * d
````

````python
a = 16
b = 4
c = 9
d = 3

result = calculation(a,b,c,d)
print(result)
````

4. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `multipy_numbers`, —è–∫–∞ –ø—Ä–∏–π–º–∞—î –¥–≤–∞ —á–∏—Å–ª–∞ —è–∫ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —ñ –ø–æ–≤–µ—Ä—Ç–∞—î —ó—Ö –¥–æ–±—É—Ç–æ–∫.

  –í–∏–∫–ª–∏—á—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `multipy_numbers` –Ω–∞ –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª
- 10 —ñ 2
- 7 —ñ 8
- 1024 —ñ 999.
  
  –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–æ–∂–Ω–æ–≥–æ –≤–∏–∫–ª–∏–∫—É.
  –ü—Ä–∏ –≤–∏–∫–æ–Ω–∞–Ω–Ω—ñ —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º —Ü–∏–∫–ª—É - –æ—Ç—Ä–∏–º–∞—î—Ç–µ 2—Ö –±–∞–ª—ñ–≤ –∑–∞ –Ω—å–æ–≥–æ :)

````python
def multipy_numbers(x, y):
    return x * y

compl_arg = [(10, 2), (7, 8), (1024, 999)]
for x, y in compl_arg:
  result = multipy_numbers(x, y)
  print(result)
````

5. **–î—ñ–ª–µ–Ω–Ω—è –∑ –ø–µ—Ä–µ–≤—ñ—Ä–∫–æ—é —Ç–∏–ø—ñ–≤.**  

  1. –ù–∞–ø–∏—à—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é `int_division` —è–∫–∞ –ø—Ä–∏–π–º–∞—î –Ω–∞ –≤—Ö—ñ–¥ –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∏ —Ç–∏–ø—É `int` - `a` —ñ `b`, —ñ –≤–∏–∫–æ–Ω—É—î –¥—ñ–ª–µ–Ω–Ω—è –ø–µ—Ä—à–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –Ω–∞ –¥—Ä—É–≥–∏–π: `a/b`. –ü—Ä–∏ —Ü—å–æ–º—É —Ñ—É–Ω–∫—Ü—ñ—è –ø–µ—Ä–µ–¥ –¥—ñ–ª–µ–Ω–Ω—è–º –ø–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ –¥—Ä—É–≥–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ –¥–æ—Ä—ñ–≤–Ω—é—î 0. –Ø–∫—â–æ –¥—Ä—É–≥–∏–π –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–µ 0, —Ñ—É–∫–Ω—Ü—ñ—è –≤–µ—Ä—Ç–∞—î —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥—ñ–ª–µ–Ω–Ω—è, —ñ–Ω–∞–∫—à–µ - –∑–∞–º—ñ—Å—Ç—å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –¥—ñ–ª–µ–Ω–Ω—è, –≤–µ—Ä—Ç–∞—î –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –Ω–∞ –µ–∫—Ä–∞–Ω "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0".

    –Ø–∫—â–æ —Ç–∏–ø–∏, –Ω–∞–¥–∞–Ω—ñ —É —Ñ—É–Ω–∫—Ü—ñ—é, –Ω–µ `int` - —Ñ—É–Ω–∫—Ü—ñ—è –≤–∏–≤–æ–¥–∏—Ç—å –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö.".

  2. –ó–∞–ø—É—Å—Ç—ñ—Ç—å —Ñ—É–Ω–∫—Ü—ñ—é –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–∏—Ö –ø–∞—Ä–∞—Ö —á–∏—Å–µ–ª:
- 12 —ñ 6
- 20 —ñ 0
- 112 —ñ 10
- 99 —ñ 0
- 11 i 1.1

    —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è "–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: " - —ñ –¥–∞–ª—ñ —Ç–µ, —â–æ –ø–æ–≤–µ—Ä–Ω—É–ª–∞ —Ñ—É–Ω–∫—Ü—ñ—è.

  3. –î–∞–π—Ç–µ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: "–ß–æ–º—É –≤ –¥–µ—è–∫–∏—Ö –≤–∏–ø–∞–¥–∫–∞—Ö —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ–≤–µ—Ä–Ω—É–ª–∞ None?"

````python
def int_division(a:int, b:int):
    if isinstance(a, int) and isinstance(b, int):
      if b != 0:
        return a / b
      else:
        return "–ù–µ –º–æ–∂—É –≤–∏–∫–æ–Ω–∞—Ç–∏ –¥—ñ–ª–µ–Ω–Ω—è –Ω–∞ 0"
    else:
      return"–ù–µ –ø—Ä–∞—Ü—é—é –∑ —Ç–∞–∫–∏–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–∏—Ö."

compl_arg2 = [(12, 6), (20, 0), (112, 10), (99, 0), (11, 1.1)]
for a, b in compl_arg2:
    result = int_division(a, b)
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–æ–±–æ—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—ó: {result}")
````
